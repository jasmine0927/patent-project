```python
import fasttext
import pandas as pd
import numpy as np
from numpy.linalg import norm
```

```python
# 加載 FastText 模型
model = fasttext.load_model(r"C:\Users\nicon\Downloads\cc.zh.300.bin\cc.zh.300.bin")
```

```python
# 讀取專利草稿構面和專利構面資料
draft_df = pd.read_excel('專利草稿構面.xlsx')
patents_df = pd.read_excel('專利構面.xlsx')
```

```python
# 欄位名稱列表
columns = ['技術特徵', '專利範圍', '獨立與從屬權利要求', '技術問題與解決方案', '摘要和背景資訊', '結構和功能元素','IPC或CPC分類'] 
```

```python
# 合併草稿構面的文本
draft_text = ' '.join(draft_df[col].iloc[0] for col in columns)
draft_vector = np.squeeze(np.asarray(model.get_sentence_vector(draft_text)))
```

```python
# 計算相似度並存儲結果
similarities = []
for idx, row in patents_df.iterrows():
    patent_text = ' '.join(str(row[col]) for col in columns)
    patent_vector = np.squeeze(np.asarray(model.get_sentence_vector(patent_text)))
    similarity = np.dot(draft_vector, patent_vector) / (norm(draft_vector) * norm(patent_vector))
    similarities.append((idx, similarity))
```

```python
# 按相似度排序並輸出結果
similarities.sort(key=lambda x: x[1], reverse=True)
for idx, similarity in similarities:
    print(f"Index: {idx+1}, Similarity: {similarity}")
```

```python
# 將結果輸出至 Excel，將索引值加 1
similarity_df = pd.DataFrame(similarities, columns=['Index', 'Similarity'])
similarity_df['Index'] = similarity_df['Index'] + 1
similarity_df.to_excel('fasttext整體_similarity_results.xlsx', index=False)

print("相似度結果已輸出至 fasttext_similarity_results.xlsx")
```
