```python
import spacy
import pandas as pd
from spacy.lang.zh.examples import sentences 
```

```python
# 加載中文模型
nlp = spacy.load("zh_core_web_lg")
```

```python
# 讀取專利草稿構面和專利構面資料
draft_df = pd.read_excel('專利草稿構面.xlsx')
patents_df = pd.read_excel('專利構面.xlsx')
```

```python
# 只處理“專利範圍”欄位
draft_text = draft_df['專利範圍'].iloc[0]
draft_doc = nlp(draft_text)
```

```python
# 計算相似度並存儲結果
similarities = []
for idx, row in patents_df.iterrows():
    patent_text = str(row['專利範圍'])
    patent_doc = nlp(patent_text)
    similarity = draft_doc.similarity(patent_doc)
    similarities.append((idx, similarity))
```

```python
# 按相似度排序並輸出結果
similarities.sort(key=lambda x: x[1], reverse=True)
for idx, similarity in similarities:
    print(f"Index: {idx+1}, Similarity: {similarity}")
```

```python
# 將結果輸出至 Excel，將索引值加 1
similarity_df = pd.DataFrame(similarities, columns=['Index', 'Similarity'])
similarity_df['Index'] = similarity_df['Index'] + 1
similarity_df.to_excel('spacy專利範圍_similarity_results.xlsx', index=False)

print("相似度結果已輸出至 spacy_similarity_results.xlsx")
```
