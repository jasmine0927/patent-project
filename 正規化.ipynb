```python
import pandas as pd
```

```python
# 讀取 Excel 檔案
file_path = 'GPT構面比對原始4.xlsx'  # 請替換成您的檔案路徑
df = pd.read_excel(file_path)
```

```python
# 選擇要正規化的欄位，假設從第二欄開始到第九欄
columns_to_normalize = df.columns[1:9]  # 調整這裡以符合實際的欄位位置
```

```python
# 進行正規化處理
for column in columns_to_normalize:
    max_value = df[column].max()
    min_value = df[column].min()
    df[column] = (df[column] - min_value) / (max_value - min_value)
```

```python
# 將處理後的 DataFrame 存回 Excel 檔案
output_file_path = 'GPT構面正規化4.xlsx'  # 請替換成您希望存儲的檔案路徑
df.to_excel(output_file_path, index=False)
```
